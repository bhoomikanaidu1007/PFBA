```python
product = input("Product name: ")
price = float(input("Unit price (₹): "))
qty = int(input("Quantity: "))
discount = float(input("Discount (%): "))
total_before = price * qty
discount_amt = total_before * discount / 100
total_after = total_before - discount_amt
print()
print("===== INVOICE =====")
print(f"Product : {product}")
print(f"Quantity : {qty}")
print(f"Price : ₹{price}")

print(f"Discount : {discount}% (-₹{discount_amt:.2f})")
print(f"TOTAL : ₹{total_after:.2f}")
print("===================")
```

```python
prices = [500, 3500, 12000, 850, 7800]
for price in prices:
    if price < 1000:
        print(f"₹{price} → Budget")
    elif price <= 10000:
        print(f"₹{price} → Mid-range")
    else:
        print(f"₹{price} → Premium")
```
