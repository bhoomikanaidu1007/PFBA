```python
import pandas as pd

df = pd.read_csv('sales_data_raw.csv')
print(df)

print("=" * 45)
print(f"Rows & Columns     : {df.shape[0]} rows, {df.shape[1]} columns")

missing = df.isnull().sum()
most_missing_col = missing.idxmax()
most_missing_val = missing.max()
print(f"Most Missing Values: '{most_missing_col}' ({most_missing_val} missing)")

if "Unit_Price" in df.columns:
    avg_price = df["Unit_Price"].mean()
    print(f"Avg Unit_Price     : {avg_price:.2f}")
else:
    print("Column 'Unit_Price' not found — check your column name.")

duplicate_count = df.duplicated().sum()
print(f"Duplicate Rows     : {duplicate_count}")
print("=" * 45)
```
